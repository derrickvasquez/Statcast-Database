```python
import pandas as pd
import csv
```

```python
# get data for pitchers with a minimum of 200 pitches thrown in 2022

from pybaseball import statcast_pitcher_pitch_arsenal

data_pitcher_indvidual = statcast_pitcher_pitch_arsenal(2022, minP=200, arsenal_type="avg_speed")
df_data_pitcher_indvidual = pd.DataFrame(data_pitcher_indvidual)
df_data_pitcher_indvidual
```

```python
# get data for batters with a minimum of 100 batted balls in 2022

from pybaseball import statcast_batter_exitvelo_barrels

data_hitter_individual = statcast_batter_exitvelo_barrels(2022, 100)
df_data_hitter_indvidual = pd.DataFrame(data_hitter_individual)
df_data_hitter_indvidual
```

```python
# Creating CSV Files

df_data_pitcher_indvidual.to_csv('pitcher_individual.csv', index=True, header=True)
df_data_hitter_indvidual.to_csv('hitter_individual.csv', index=True, header=True)
```

# In 2022, who had the fastest average fastball? 

```python
%%sql

select first_name, last_name, ff_avg_speed,
RANK() OVER (ORDER BY ff_avg_speed DESC)
from ind_pitcher_statcast
where ff_avg_speed is not null
limit 10;
```

## Using the RANK() function within SQL, we can create a ranking query to find who had the 10 average fastest fourseam fastball velocities 

### From our query, we can see Jhoan Duran from the Twins held the fastest average fastball in 2022 WITH 100.8 MPH!

# In 2022, who had the fastest average exit velocity? 

```python
%%sql

select first_name, last_name, attempts, avg_hit_speed,
RANK() OVER (ORDER BY avg_hit_speed DESC)
from ind_hitter_statcast
where avg_hit_speed is not null
limit 10;
```

### From our query, we can see Aaron Judge from the Yankees, and AL MVP, held the hardest average exit velocity in 2022 with 95.9 MPH!

# In 2022, who had the slowest average fastball?

```python
%%sql

select first_name, last_name, ff_avg_speed,
RANK() OVER (ORDER BY ff_avg_speed asc)
from ind_pitcher_statcast
where ff_avg_speed is not null
limit 10;
```

### From our query, we can see Tyler Rogers from the Giants, held the slowest average fastball velocity in 2022 with 83.2 MPH. However, this data can be misinterpreted to those who do not follow baseball closely. 

### Rogers is known as a 'Submarine' pitcher, meaning he throws from an unusually low arm slot. With that in mind, his 83.2 MPH fastball plays much faster in game-speed and causes a lot of havoc to hitters who have to face him!

# In 2022, who had the slowest average exit velocity?

```python
%%sql

select first_name, last_name, attempts, avg_hit_speed,
RANK() OVER (ORDER BY avg_hit_speed asc)
from ind_hitter_statcast
where avg_hit_speed is not null
limit 10;
```

### From our query, David Fletcher from the Angels held the lowest average exit velocity in 2022. With some further research, it is found Fletcher did not play in many games this season due to injury. This might be the culprit for his loss of exit velocity.

# Now, I want to find the ev95percent of all players who had over 300 attempts


```python
%%sql

select first_name, last_name, attempts, ev95percent
from ind_hitter_statcast
where avg_hit_speed is not null and attempts > 300
order by ev95percent desc;
```

# Now, I want to find the average hit angle in comparison to average hit distance/speed of all players who had over 300 attempts


```python
%%sql

select first_name, last_name, attempts, avg_hit_angle, avg_distance, avg_hit_speed
from ind_hitter_statcast
where attempts > 300
order by avg_hit_speed desc;
```

## avg_hit_angle is simply Launch Angle, which represents the vertical angle at which the ball leaves a player's bat after being struck.

## The greater the launch angle, the more lift the ball has. The lower the angle, the more chance of the ball being classified as a ground ball.
